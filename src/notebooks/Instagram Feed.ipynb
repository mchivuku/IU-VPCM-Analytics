## Retrieving Instagram Feed

```python
from IPython.display import display, Image
import pprint
response = requests.get("https://api.instagram.com/v1/users/self/media/recent/?access_token="+ACCESS_TOKEN)
recent_posts = response.json()
```

```python
def display_image_feed(feed, include_caption = True):
    for post in feed['data']:
        display(Image(url=post["images"]["low_resolution"]["url"]))
        print(post["images"]["standard_resolution"]["url"])
        if include_caption:
            print(post["caption"]["text"])
        print()
```

```python
display_image_feed(recent_posts)
```

```python
import json

print(json.dumps(recent_posts, indent=2))
```
